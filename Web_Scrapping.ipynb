```python
house_type = []  # house type i.e 1,2,3 bhk
location = []   # house location brookiefiled,hsr layout
price = [] #price per each flat
builtup_area = []
furnishing_status =[]
Highlights = []
agent_name = []
verification_status = []
count_of_buyers_contacted = []


for i in range(1,15):
    housing_url = f'https://housing.com/rent/flats-for-rent-in-bengaluru-karnataka-P38f9yfbk7p3m2h1f?page={i}'
    housing_link = requests.get(housing_url)
    print(housing_link.status_code)

    housing_soup = BeautifulSoup(housing_link.text, 'html.parser')

    
    for i in housing_soup.find_all('h2',class_ = 'title-style'):
        full_detail = i.text
        house_type.append(i.text.split('for rent in')[0].strip())
        location.append(i.text.split('for rent in')[-1].strip())

    for i in housing_soup.find_all('div',class_ = "T_singlePriceStyle _7l1ulh _r31e5h _g3gktf _csbfng _c8exct _bx1t02"):
        price.append(i.text)

    
    for i in housing_soup.find_all('div', class_ = "T_primaryInfoTextStyle _t9glyw _c81fwx T_descriptionStyle _7l1472 _r31e5h _g3exct _csbfng _bx1t02"):
        text= i.text
        built_up = re.findall(r'\d+\ssq.ft',text,re.IGNORECASE)
        if built_up:
            builtup_area.append(text)
        else:
            furnishing_status.append(text)

    for i in housing_soup.find_all('div', class_ = '_c81fwx _cs1nn1 _vy1wug _g3gktf _1qkcs3je T_dataPointStyle'):
        high_text = i.text #'Highlights: 24x7 Security \xa0•\xa0 Swimming Pool \xa0•\xa0 Children Play Area \xa0•\xa0 Close to Hospital'
        high_split = high_text.split('Highlights:')[-1].strip() #'24x7 Security \xa0•\xa0 Swimming Pool \xa0•\xa0 Children Play Area \xa0•\xa0 Close to Hospital'
        high_split = high_split.split('•')
        updated_highlights = []
        for j in high_split:
            j.replace('\xa0','')
            updated_highlights.append(j.strip())
        Highlights.append(updated_highlights)

    for i in housing_soup.find_all('div', class_ = "sellerName-label"):
        agent_name.append(i.text.strip())


    for i in housing_soup.find_all('div', class_ = 'sellerName'):
        finding_count = re.findall(r'\d+.\sbuyers',i.text)
        if finding_count:
            for j in housing_soup.find_all('div',class_ = 'buyers-served text T_contactedBuyersStyle _xss319bv _12un19bv _0h1h6o _cx1y44 _9s1txw _7l9ke2 _fr1ti3 _g31fwx _cs1nn1 _c819bv _bx1t02'):
                count = j.text
                if count:
                    count = count.split(' buyers')
                    count_of_buyers_contacted.append(count[0].strip())
                    break
        else:
            count_of_buyers_contacted.append(np.nan)

    for j in housing_soup.find_all('div',class_ = 'srp-tags-container T_tagContainerStyle _18ibftgi _fqexct _tkftgi _cx1y44 _mkstnw _9s1txw'):
        verify_text = re.findall('Verified',j.text, re.IGNORECASE)
        if verify_text:
            if len(j.text) > len('Verified'):
                verification_status.append(j.text[:8] + "," + j.text[8:])
            else:
                verification_status.append(j.text)
        else:
            verification_status.append(j.text)

    time.sleep(random.uniform(5,10))
```

```python
data = {'House Type':house_type,'Location':location,'Price':price,'Built up Area': builtup_area,'Furnishing Status':furnishing_status,'Highlights/Insights': Highlights,'Agent Name':agent_name,'Verification Status':verification_status, 'No. of customer Contacted':count_of_buyers_contacted}
```
