## Problem Statement : Availability of rental properties by location in Hyderabad

#### Univariate Analysis

    --> Getting Information about availability of rental properties based on locations.. 

```python
unique_location_counts = hyderabad_rental_df['Location'].value_counts()
unique_location_counts_df = pd.DataFrame(unique_location_counts)
```

```python
#Bar Chart: Availability of rental properties by location in Hyderabad
plt.figure(figsize= (15,6))
plot_ = sns.barplot(x = unique_location_counts.index,y = unique_location_counts.values)
plt.xticks(rotation = 90)

for i, value in enumerate(unique_location_counts.values):
    plot_.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=6, rotation=90)


plt.title('Chart on Availability of rental properties by location in Hyderabad')
plt.xlabel('Different Location Names in Hyderabad City--->')
plt.ylabel('Count of rental properties available---->')
#plt.tight_layout()
plt.grid(axis= 'y')
plt.show()
```

#### Bivariate Analysis:

    --> Comparing the Price and location to justify the available location counts..

```python
#Comparing the locations based upon price range of top 5 available locations and last 5 available locations:

# Top 5 location:
top_5_locations = unique_location_counts.head(5).index
last_6_locations = unique_location_counts.tail(6).index

top_5_df = hyderabad_rental_df[hyderabad_rental_df['Location'].isin(top_5_locations)]
top_5_df['Position In'] = 'Most Available locations'
last_6_df = hyderabad_rental_df[hyderabad_rental_df['Location'].isin(last_6_locations)]
last_6_df['Position In'] = 'Less Available Property Locations'
```

```python
top_5_df['Location'].value_counts()
```

```python
total_filtered_locations = pd.concat([top_5_df,last_6_df])
# total_filtered_locations.head()
```

```python
plt.figure(figsize= (16,6))
analysis_plot = sns.barplot(data = total_filtered_locations,x = 'Location', y = 'Price', hue = 'Position In', errorbar= None)

plt.title('Analysis Between Price and Location to check Availability in Top and Last 5 Locations', fontweight = 'bold', fontsize = 14, pad = 20)
plt.xlabel('Location Name', fontweight = 'bold', fontsize = 12)
plt.ylabel('Price of Rental Property--->', fontweight = 'bold', fontsize = 12)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
## Analysing Based on above Graph:
print('Average Price in most available locations:',top_5_df['Price'].mean())
print('Average Price in less available locations:', last_6_df['Price'].mean())
```

##### Analysis Based on above plot

    ---> the Average Price in Most available locations is '62333.80'
    --> The Average Price in less available locations is '124333.33'

Which clearly tells that Most Costly area are having less availability of rental properties and Less costly and budget friendly properties are having more available rental properties in hyderabad city 

#### Multivariate Analysis

    --> Comparing the Price and location and Justifying through Property Type such as Villa's or Apartments or Independent Houses

```python
hyderabad_rental_df['House Type'].unique()
```

```python
def simplify_house_type(house_type):
    if 'Flat' in house_type:
        return 'Flat'
    elif 'Villa' in house_type:
        return 'Villa'
    elif 'Independent House' in house_type:
        return 'Independent House'
    elif 'Builder Floor' in house_type:
        return 'Builder Floor'
    else:
        return 'Other'
```

```python
total_filtered_locations['Grouped Type'] = total_filtered_locations['House Type'].apply(simplify_house_type)

# Step 2: Plotting
plt.figure(figsize=(10,5))
sns.barplot(data=total_filtered_locations,x='Location',y='Price',hue='Grouped Type', ci = None)

plt.title('Multivariate Analysis: Price vs Location by Grouped House Type',fontweight = 'bold',fontsize = 14)
plt.xlabel('Location',fontweight = 'bold',fontsize = 12)
plt.ylabel('Price--->',fontweight = 'bold',fontsize = 12)
plt.xticks(rotation=90)
plt.legend(title='House Type')
plt.tight_layout()
plt.show()
```

##### Analysis Based on above plot
    Getting the location names and analysing the price pattern at different locations and different properties..

    ---> Here we can see that the villas at more available locaitons also are at high cost and where as when we are comparing the less available locations most of them are villa's and Independent Houses..

#### Heat Map

```python
heatmap_data = total_filtered_locations.pivot_table(index='Location',columns='House Type',values='Price',aggfunc='mean')

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data,annot=True,fmt=".0f",cmap="coolwarm",linewidths=0.5,linecolor='gray')

plt.title('Average Rental Price by Location and House Type', fontweight='bold',fontdict={'family': 'Times New Roman', 'fontsize': 14})
plt.xlabel('House Type', fontweight='bold')
plt.ylabel('Location', fontweight='bold')
plt.xticks(rotation=75)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```


    --> This heat map is used when we want to get a house for a rent with less price despite of any location we can use this heat map and we can get the correct flat or villa at available prices

    --> We can use this Heat map to get a rental property at less price comparitively depite of location.


## Problem Statement : Price and availability analysis on Rental properties city wise

```python
df = pd.read_csv('Cleaned Rental Data.csv')
```

```python
df.columns = df.columns.str.strip()
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['No. of customer Contacted'] = pd.to_numeric(df['No. of customer Contacted'], errors='coerce')
df_clean = df.dropna(subset=['Price', 'No. of customer Contacted'])
```

#### Univariate Analysis

```python
sns.histplot(df_clean['Price'], bins=30, kde=True, color = '#ce4800')
plt.title('Distribution of Rental Price',fontweight = 'bold')
plt.xlabel('Price',fontweight = 'bold')
plt.ylabel('Frequency',fontweight = 'bold')
plt.show()
```

##### Analysis on the above images :

    --> As we can check in the image there are many outliers in the price of the buildings as per the house type and locality type the price is varing in the cities.

```python
## Finding types of furnishing status Available as per count of the type 
sns.countplot(data=df_clean, x='Furnishing Status', order=df_clean['Furnishing Status'].value_counts().index, palette='muted')
plt.title('Furnishing Status Distribution',fontweight = 'bold')
plt.show()
```

##### Furnishing Status of the Images:

    --> Many of the rental properties available are semi furnished and un-furnished rental properties are very less when compared to the semi furnished and fully furnished 

```python
##Frequency of type of houses available

plt.figure(figsize=(12,7))
sns.countplot(data=df_clean, x='House Type', order=df_clean['House Type'].value_counts().index, palette='viridis')
plt.title('Frequency of House Types',fontweight = 'bold')
plt.xlabel('Types of Houses Available', fontweight = 'bold')
plt.ylabel('Count of Available houses Present --> ',fontweight = 'bold')
plt.xticks(rotation=90)
plt.show()
```

##### 
        --> As we can check here most of the available properties in different cities are Flats rather than independet houses or villa's in a locality

```python
### pie chart for checking the percentage of Verified properties:
df_clean['Verification Status'] = df_clean['Verification Status'].replace('Verified,New','Verified')
verification_counts = df_clean['Verification Status'].value_counts()

# Plot pie chart
plt.figure(figsize=(5,5))
plt.pie(verification_counts,labels=verification_counts.index,autopct='%1.1f%%',startangle=90,wedgeprops={'edgecolor': 'black'})

plt.title('Showing Details about verfied properties',pad = 20)
plt.axis('equal')
plt.tight_layout()
    
plt.show()
```

```python
##Comparision between furnishing status and number of customers contacted:

plt.figure(figsize=(10,6))
sns.boxplot(data=df_clean, x='Furnishing Status', y='No. of customer Contacted', palette='muted')
plt.title('Comparison Between Furnishing Status and Number of Customers Contacted')
plt.xlabel('Furnishing Status')
plt.ylabel('Number of Customers Contacted')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

##### Based upon the boxplots given 

    --> Semi furnished properties receive the highest and most varied customer interest, indicating strong overall demand. 
    --> Fully furnished and un-furnished properties attract fewer contacts on average, with un-furnished showing occasional spikes due to some customer long-term renters, or budget-conscious buyers who want to customize the space themselves .
    --> This suggests prioritizing semi furnished listings for broader market engagement.

```python
## Price Distribution across the citites..

plt.figure(figsize=(8, 4))
sns.violinplot(data=df_clean, x='City', y='Price', palette='bright')

plt.title('Price Distribution Across Cities', fontsize=16,fontweight = 'bold')
plt.xlabel('City',fontweight = 'bold' ,fontsize=12,)
plt.ylabel('Price',fontweight = 'bold', fontsize=12,)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='dotted', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### Analysis on the above plot 

    --> Madras exhibits the widest spread, suggesting a highly diverse property market from low to very high prices.
    --> Calcutta has a narrower range and fewer high-end outliers, suggesting more price consistency.
    --> Madras leads with the highest and most varied property prices, indicating a wide-ranging market including luxury segments. Calcutta offers the most consistent and affordable pricing, making it attractive for budget-conscious buyers.

#### Bi variate Analysis

```python
## Comparing Price and area by furnishing status
plt.figure(figsize=(6,4))
sns.scatterplot(data=df_clean,x='Built up Area in sq.ft',y='Price',hue='Furnishing Status',alpha=0.7,palette='Dark2')

plt.title('Price vs Built-up Area by Furnishing Status', fontsize=12,fontweight = 'bold')
plt.xlabel('Built-up Area (sq.ft)', fontsize=10,fontweight = 'bold')
plt.ylabel('Price', fontsize=10,fontweight = 'bold')
plt.legend(title='Furnishing')
plt.tight_layout()
plt.show()
```

##### Analysis bases on the above Image

    --> There is a clear positive correlation between built-up area and price  larger properties tend to be more expensive.
    --> Theres a strong positive correlation between built-up area and price, with fully furnished properties commanding higher prices overall. Semi furnished units dominate the mid-range, while un-furnished homes mostly remain in the budget segment.

#### Multivariate Analysis

```python
## Getting relation between Price , built up area and number of customers contacted

plt.figure(figsize=(8,6))
sns.heatmap(df_clean[['Price','Built up Area in sq.ft','No. of customer Contacted']].corr(), annot=True, cmap='vlag',annot_kws={"size": 12, "color": "black"})
plt.title('Correlation Heatmap between Price, Area And Number of customer contacted', fontsize=12,fontweight = 'bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

##### Anlaysis based on above plot:

    --> Price and Built-up Area have a strong positive correlation of 0.81, indicating that larger homes tend to be priced higher.
    --> Price and Number of Customers Contacted (0.08) and Area and Number Contacted (0.11) show very weak correlations, implying customer engagement isn't influenced much by just size or price.

    --> Price strongly correlates with built-up area (r = 0.81), suggesting bigger homes cost more. However, customer interest (contact rate) is weakly related to both area and price.
